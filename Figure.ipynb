```python
labels = 'accident', 'art', 'crime', 'economics', 'education', 'entertainment', 'environment', 'international', 'opinion', 'politics', 'science_tech', 'sports'
sizes = [5036, 2117, 6858, 2731, 9646, 7913, 3439, 4121, 6430, 15867, 2315, 9478]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels = labels, autopct = '%1.1f%%', pctdistance = 0.82, startangle = 180, labeldistance = 1.07)
ax1.axis('equal')
plt.title("Class distribution")
plt.tight_layout()
plt.savefig('pie', dpi = 500)
plt.show()
```

```python
accu = {'RandomForest':0.68, 'MultinomialNB':0.72, 'LogisticRegression':0.77, 'MLP':0.77, 'XGBoost':0.77, 'SVM':0.78 , 'LSTM':0.83}
classifiers = list(accu.keys())
values = list(accu.values())
fig, ax5 = plt.subplots(figsize =(6, 3))
b = plt.barh(classifiers, values, color ='seagreen', height = 0.5)
plt.ylabel("Classifiers")
plt.xlabel("Accuracy Score")
plt.title("Accuracy Comparison")
ax5.bar_label(b, padding = -25)
plt.tight_layout()
plt.savefig('accuracy', dpi = 500)
plt.show()
```

```python
pre = {'RandomForest':0.68, 'MultinomialNB':0.72, 'LogisticRegression':0.76, 'MLP':0.77, 'XGBoost':0.77, 'SVM':0.78 , 'LSTM':0.83}
classifier = list(pre.keys())
valuesp = list(pre.values())
fig, ax6 = plt.subplots(figsize =(6, 3))
c = plt.barh(classifier, valuesp, color ='dodgerblue', height = 0.5)
plt.ylabel("Classifiers")
plt.xlabel("Precision Score")
plt.title("Precision Comparison")
ax6.bar_label(c, padding = -25)
plt.tight_layout()
plt.savefig('precision', dpi = 500)
plt.show()
```

```python
f1 = {'RandomForest':0.67, 'MultinomialNB':0.71, 'LogisticRegression':0.76, 'MLP':0.76, 'XGBoost':0.77, 'SVM':0.78 , 'LSTM':0.83}
classifierf = list(f1.keys())
valuesf = list(f1.values())
fig, ax8 = plt.subplots(figsize =(6, 3))
e = plt.barh(classifierf, valuesf, color ='slategray', height = 0.5)
plt.ylabel("Classifiers")
plt.xlabel("F1-Score")
plt.title("F1-Score Comparison")
ax8.bar_label(e, label_type='center')
plt.tight_layout()
plt.savefig('f1', dpi = 500)
plt.show()
```

```python
a1 = [0.55, 0.58, 0.69, 0.58]
a2 = [0.71, 0.72, 0.72, 0.72]

fig, ax = plt.subplots(figsize =(8, 6))
bar_width = 0.2
X = np.arange(4)

p2 = plt.barh(X + bar_width, a2, bar_width, color = 'darkgreen', label = 'alpha=0.01')
p1 = plt.barh(X, a1, bar_width, color = 'limegreen', label = 'alpha=1.0')

plt.ylabel('MultinomialNB')
plt.xlabel('Score')
plt.title('Alpha Value Comparison for MultinomialNB')
plt.yticks(X + (bar_width/2) , ('F1-Score', 'Recall', 'Precision', 'Accuracy'))
plt.legend()
ax.bar_label(p2)
ax.bar_label(p1)
plt.tight_layout()
plt.savefig('Alpha', dpi = 500)
plt.show()
```

```python
b1 = [0.76, 0.77, 0.77, 0.77]
b2 = [0.71, 0.71, 0.71, 0.71]

fig3, ax3 = plt.subplots(figsize =(6, 6))
bar_width = 0.2
X3 = np.arange(4)

p2 = plt.barh(X3 + bar_width, b1, bar_width, color = 'goldenrod', label = 'With EarlyStopping')
p1 = plt.barh(X3, b2, bar_width, color = 'dimgrey', label = 'Without EarlyStopping')

plt.ylabel('MLP')
plt.xlabel('Score')
plt.title('EarlyStopping for MLP')
plt.yticks(X2 + (bar_width/2) , ('F1-Score', 'Recall', 'Precision', 'Accuracy'))
plt.legend()
ax3.bar_label(p2, padding = -25)
ax3.bar_label(p1, padding = -25)
plt.tight_layout()
plt.savefig('earlyStop_MLP', dpi = 500)
plt.show()
```
